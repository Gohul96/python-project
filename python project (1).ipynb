```python
#Project
#Salary Slip for the employee
#Basic Pay
#HRA--> house rent allowances---10%
#MA---> Medical Allowances----5%
#Spouse Allowances----2%

#Tax
# 1) Professional Tax ----500
# 2) Income Tax
#    a) <25000 -- 0
#    b) 25000< and <50000 -- 5%
#    c) 50000< and <100000---10%
#    d) > 1 lac ---25%

#insurance
#3400 if your age >50
#2600 if your age <50

#if i input n persons name, salary and age i should get n outputs

#Input of Name and Basic pay,age
#create a function which basically takes name and basic pay as the input and returns the net salary and also gross salary
#Basic Pay+HRA+MA+Spouse Allowance-------> Net salary
#Deductions= net salary-P.Tax-IT-Insurance

name=input("employee's name:")
age=int(input("employee's age:"))
basic_pay=int(input("employee's basic pay:"))
house_rent_allowances=basic_pay*0.1
medical_allowances=basic_pay*0.05
spouce_allowances=basic_pay*0.02
allowances=house_rent_allowances + medical_allowances + spouce_allowances

professional_tax=500
income_tax=basic_pay
if basic_pay<=25000:
    income_tax=0
elif basic_pay>25000 and basic_pay<50000:
    income_tax=basic_pay*0.05
elif basic_pay>50000 and basic_pay<100000:
    income_tax=basic_pay*0.1
elif basic_pay>=100000:
    income_tax=basic_pay*0.25


insurance=age
if age<50:
    insurance=2600
else:
    insurance=3400

print("Hi",name)
print("age:",age)
print("basic pay:",basic_pay)
print("*************************************")
print("                    EMPLOYEE'S SALARY SLIP")
print("************************************")
print("ALLOWANCES")
print("house rent allowances:",house_rent_allowances)
print("medical allowances:",medical_allowances)
print("spouce allownaces:",spouce_allowances)

print("************************************")
print("TAXES:")
print("professional tax",professional_tax)
print("income_tax",income_tax )
print("************************************")
print("INSURANCE")
print("insurance:",insurance)

total_salary = basic_pay-allowances
Net_salary=total_salary+(professional_tax-income_tax)
print("NET SALARY:",Net_salary)
print("*********************THANK YOU********************")

```
